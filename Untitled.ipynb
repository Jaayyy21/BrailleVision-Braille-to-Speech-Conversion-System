```python
import torch, os, json, random, shutil
import cv2
import matplotlib.pyplot as plt
from ultralytics import YOLO
from gtts import gTTS
```

```python
PROJECT_ROOT = os.getcwd() # Gets the current working directory of the notebook

VAL_SPLIT_RATIO = 0.2
RAW_DATA_DIR = os.path.join(PROJECT_ROOT, "raw_dataset")
RAW_IMAGES_DIR = os.path.join(RAW_DATA_DIR, "images")
RAW_ANNOTATIONS_DIR = os.path.join(RAW_DATA_DIR, "annotations")
YOLO_DATASET_DIR = os.path.join(PROJECT_ROOT, "braille_yolo_dataset")
TEST_IMAGE_PATH = os.path.join(PROJECT_ROOT, "test_images/sample.jpg")

# Ensure the main project folder exists (though it should, since you're running the notebook here)
os.makedirs(PROJECT_ROOT, exist_ok=True)
print(f"Working directory (PROJECT_ROOT) set to: {PROJECT_ROOT}")
```

```python
print("\nChecking for CUDA availability...")
DEVICE = 'cuda' if torch.cuda.is_available() else 'cpu'
print(f"Using device: {DEVICE}")
if DEVICE == 'cuda':
    print(f"CUDA Version: {torch.version.cuda}")
    print(f"GPU Name: {torch.cuda.get_device_name(0)}")
```

```python
print("\n Starting automated dataset preparation...")

# 2.1 Find Unique Labels & Create CLASS_MAP
unique_labels = set()
if os.path.isdir(RAW_ANNOTATIONS_DIR):
    for filename in os.listdir(RAW_ANNOTATIONS_DIR):
        if filename.endswith(".json"):
            try:
                with open(os.path.join(RAW_ANNOTATIONS_DIR, filename)) as f:
                    data = json.load(f)
                for shape in data.get('shapes', []):
                    if 'label' in shape:
                        unique_labels.add(shape['label'])
            except: 
                continue

sorted_labels = sorted(list(unique_labels))
CLASS_MAP = {label: i for i, label in enumerate(sorted_labels)}
print(f"   - Found {len(CLASS_MAP)} unique classes.")
```

```python
# 2.2 Conversion Function (Kept as is)
def convert_bbox_to_yolo(img_width, img_height, class_id, bbox_points):
    x_min, y_min = bbox_points[0]
    x_max, y_max = bbox_points[1]
    width, height = x_max - x_min, y_max - y_min
    x_center = (x_min + width / 2.0) / img_width
    y_center = (y_min + height / 2.0) / img_height
    norm_width = width / img_width
    norm_height = height / img_height
    # Ensures values are clipped between 0 and 1
    x_center = max(0, min(1, x_center))
    y_center = max(0, min(1, y_center))
    norm_width = max(0, min(1, norm_width))
    norm_height = max(0, min(1, norm_height))
    return f"{class_id} {x_center:.6f} {y_center:.6f} {norm_width:.6f} {norm_height:.6f}"
```

```python
# 2.3 Create YOLO dataset structure & Split (train/val)
if os.path.exists(YOLO_DATASET_DIR):
    shutil.rmtree(YOLO_DATASET_DIR)
for split in ['train', 'val']:# Removed 'test' split folders from creation as per original logic
    os.makedirs(os.path.join(YOLO_DATASET_DIR, 'images', split), exist_ok=True)
    os.makedirs(os.path.join(YOLO_DATASET_DIR, 'labels', split), exist_ok=True)

image_filenames = [f for f in os.listdir(RAW_IMAGES_DIR) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]
random.shuffle(image_filenames)
split_index = int(len(image_filenames) * (1 - VAL_SPLIT_RATIO))
train_files, val_files = image_filenames[:split_index], image_filenames[split_index:]
file_splits = {'train': train_files, 'val': val_files}
```

```python
# 2.4 Convert and Copy files
for split, filenames in file_splits.items():
    for img_filename in filenames:
        basename, _ = os.path.splitext(img_filename)
        json_filename = basename + '.json'
        json_path = os.path.join(RAW_ANNOTATIONS_DIR, json_filename)
        
        if not os.path.exists(json_path): continue
        
        with open(json_path, 'r') as f: data = json.load(f)
        
        # Check if image size info is available
        image_width = data.get('imageWidth', 0)
        image_height = data.get('imageHeight', 0)
        if image_width == 0 or image_height == 0:
            # Fallback: Read image to get its dimensions if not in JSON
            # NOTE: Requires OpenCV, which is imported
            img_path = os.path.join(RAW_IMAGES_DIR, img_filename)
            img = cv2.imread(img_path)
            if img is not None:
                image_height, image_width, _ = img.shape
            else:
                print(f"WARNING: Could not determine dimensions for {img_filename}. Skipping.")
                continue

        annotations = data.get('shapes', [])
        yolo_lines = []
        for ann in annotations:
            class_label = ann['label']
            if class_label in CLASS_MAP:
                # Correctly pass dimensions and points
                yolo_lines.append(convert_bbox_to_yolo(image_width, image_height, CLASS_MAP[class_label], ann['points']))
        
        if yolo_lines:
            label_path = os.path.join(YOLO_DATASET_DIR, 'labels', split, basename + '.txt')
            with open(label_path, 'w') as f: f.write('\n'.join(yolo_lines))
            # Copy image to the new YOLO structure
            shutil.copyfile(os.path.join(RAW_IMAGES_DIR, img_filename), os.path.join(YOLO_DATASET_DIR, 'images', split, img_filename))

print("Dataset preparation complete!")
```

```python
#CREATE DATA.YAML CONFIG FILE
# The path should be within the YOLO_DATASET_DIR, and paths inside the YAML 
# should be relative to the PROJECT_ROOT (as the training command is run from there)
DATA_YAML_PATH = os.path.join(PROJECT_ROOT, "braille_data.yaml") # Moved YAML to project root

# Define paths relative to the PROJECT_ROOT for the YAML file
train_path = os.path.join(YOLO_DATASET_DIR, 'images/train')
val_path = os.path.join(YOLO_DATASET_DIR, 'images/val')

with open(DATA_YAML_PATH, 'w') as f:
    f.write(f"path: {PROJECT_ROOT}\n") # Setting root path
    f.write(f"train: {os.path.relpath(train_path, PROJECT_ROOT)}\n") # Relative path
    f.write(f"val: {os.path.relpath(val_path, PROJECT_ROOT)}\n")     # Relative path
    f.write(f"nc: {len(CLASS_MAP)}\n")
    f.write(f"names: {sorted_labels}\n")

print(f" `braille_data.yaml` saved to {DATA_YAML_PATH}")
```
