```python
# ## Step 3: Configuration and Setup

# Project paths
PROJECT_ROOT = os.getcwd()
VAL_SPLIT_RATIO = 0.2

# Data directories
RAW_DATA_DIR = os.path.join(PROJECT_ROOT, "raw_dataset")
RAW_IMAGES_DIR = os.path.join(RAW_DATA_DIR, "images")
RAW_ANNOTATIONS_DIR = os.path.join(RAW_DATA_DIR, "annotations")
YOLO_DATASET_DIR = os.path.join(PROJECT_ROOT, "braille_yolo_dataset")
TEST_IMAGE_PATH = os.path.join(PROJECT_ROOT, "test_images/sample.jpg")

os.makedirs(PROJECT_ROOT, exist_ok=True)
print(f"✓ Working directory: {PROJECT_ROOT}")

# Check CUDA
DEVICE = 'cuda' if torch.cuda.is_available() else 'cpu'
print(f"\n✓ Using device: {DEVICE}")

if DEVICE == 'cuda':
    print(f"  - CUDA Version: {torch.version.cuda}")
    print(f"  - GPU Name: {torch.cuda.get_device_name(0)}")
    print(f"  - GPU Memory: {torch.cuda.get_device_properties(0).total_memory / 1e9:.2f} GB")
else:
    print("  ⚠ No GPU detected. Training will be slower on CPU.")
```

```python
# ## Step 4: Validate Dataset Directories

# %%
# Check if directories exist
if not os.path.exists(RAW_IMAGES_DIR):
    raise FileNotFoundError(f"❌ Images directory not found: {RAW_IMAGES_DIR}")
if not os.path.exists(RAW_ANNOTATIONS_DIR):
    raise FileNotFoundError(f"❌ Annotations directory not found: {RAW_ANNOTATIONS_DIR}")

image_count = len([f for f in os.listdir(RAW_IMAGES_DIR) if f.lower().endswith(('.png', '.jpg', '.jpeg'))])
annotation_count = len([f for f in os.listdir(RAW_ANNOTATIONS_DIR) if f.endswith('.json')])

print(f"✓ Found {image_count} images")
print(f"✓ Found {annotation_count} annotation files")
```

```python
 ## Step 5: Extract Unique Classes

# %%
print("Extracting unique class labels...")
unique_labels = set()
annotation_files = [f for f in os.listdir(RAW_ANNOTATIONS_DIR) if f.endswith(".json")]

for filename in annotation_files:
    try:
        with open(os.path.join(RAW_ANNOTATIONS_DIR, filename)) as f:
            data = json.load(f)
        for shape in data.get('shapes', []):
            if 'label' in shape:
                unique_labels.add(shape['label'])
    except Exception as e:
        print(f"  ⚠ Warning: Could not process {filename}: {e}")
        continue

sorted_labels = sorted(list(unique_labels))
CLASS_MAP = {label: i for i, label in enumerate(sorted_labels)}

print(f"\n✓ Found {len(CLASS_MAP)} unique classes:")
for label, idx in list(CLASS_MAP.items())[:15]:
    print(f"  {idx:2d}: {label}")
if len(CLASS_MAP) > 15:
    print(f"  ... and {len(CLASS_MAP) - 15} more")
```

```python
# ## Step 6: Define Conversion Function

# %%
def convert_bbox_to_yolo(img_width, img_height, class_id, bbox_points):
    """Convert bounding box coordinates to YOLO format"""
    x_min, y_min = bbox_points[0]
    x_max, y_max = bbox_points[1]
    
    width = x_max - x_min
    height = y_max - y_min
    
    x_center = (x_min + width / 2.0) / img_width
    y_center = (y_min + height / 2.0) / img_height
    norm_width = width / img_width
    norm_height = height / img_height
    
    # Clip values between 0 and 1
    x_center = max(0, min(1, x_center))
    y_center = max(0, min(1, y_center))
    norm_width = max(0, min(1, norm_width))
    norm_height = max(0, min(1, norm_height))
    
    return f"{class_id} {x_center:.6f} {y_center:.6f} {norm_width:.6f} {norm_height:.6f}"

print("✓ Conversion function defined")
```

```python
## Step 7: Create YOLO Dataset Structure
print("Creating YOLO dataset structure...")

# Remove existing dataset if present
if os.path.exists(YOLO_DATASET_DIR):
    shutil.rmtree(YOLO_DATASET_DIR)
    print("  - Removed existing dataset")

# Create directory structure
for split in ['train', 'val']:
    os.makedirs(os.path.join(YOLO_DATASET_DIR, 'images', split), exist_ok=True)
    os.makedirs(os.path.join(YOLO_DATASET_DIR, 'labels', split), exist_ok=True)

print("✓ Dataset structure created")
```

```python
# ## Step 8: Split Dataset into Train/Val

# %%
# Get all image files and shuffle
image_filenames = [f for f in os.listdir(RAW_IMAGES_DIR) 
                   if f.lower().endswith(('.png', '.jpg', '.jpeg'))]
random.shuffle(image_filenames)

# Split into train and validation
split_index = int(len(image_filenames) * (1 - VAL_SPLIT_RATIO))
train_files = image_filenames[:split_index]
val_files = image_filenames[split_index:]

print(f"✓ Train images: {len(train_files)}")
print(f"✓ Validation images: {len(val_files)}")

file_splits = {'train': train_files, 'val': val_files}
```

```python
# %% [markdown]
# ## Step 9: Convert, Copy, and Merge Files (Optimized for JSON/TXT Merge)

# %%
print("Converting annotations and copying files...")

train_count = 0
val_count = 0
skipped_count = 0

for split, filenames in file_splits.items():
    print(f"\nProcessing {split} set...")
    
    for idx, img_filename in enumerate(filenames):
        if (idx + 1) % 50 == 0:
            print(f"  Processed {idx + 1}/{len(filenames)} images...")
        
        basename, _ = os.path.splitext(img_filename)
        
        # Define paths for potential annotation sources/targets
        json_path = os.path.join(RAW_ANNOTATIONS_DIR, basename + '.json')
        yolo_txt_source_path = os.path.join(RAW_ANNOTATIONS_DIR, basename + '.txt') # Existing TXT annotation
        dst_yolo_path = os.path.join(YOLO_DATASET_DIR, 'labels', split, basename + '.txt')
        src_img = os.path.join(RAW_IMAGES_DIR, img_filename)
        dst_img = os.path.join(YOLO_DATASET_DIR, 'images', split, img_filename)
        
        annotated_file_path = None
        
        # --- MERGING LOGIC ---
        
        # Priority 1: Check for original JSON file and convert it
        if os.path.exists(json_path):
            try:
                # Code Block for JSON -> YOLO Conversion (Your original logic)
                with open(json_path, 'r') as f:
                    data = json.load(f)
                
                # Get image dimensions (using fallback logic)
                image_width = data.get('imageWidth', 0)
                image_height = data.get('imageHeight', 0)
                
                if image_width == 0 or image_height == 0:
                    img = cv2.imread(src_img)
                    if img is not None:
                        image_height, image_width, _ = img.shape
                    else:
                        raise FileNotFoundError("Could not read image dimensions.")
                
                # Perform JSON -> YOLO conversion
                yolo_lines = []
                for ann in data.get('shapes', []):
                    class_label = ann['label']
                    if class_label in CLASS_MAP:
                        yolo_lines.append(convert_bbox_to_yolo(
                            image_width, image_height, CLASS_MAP[class_label], ann['points']
                        ))
                
                # Save the newly converted YOLO file
                if yolo_lines:
                    with open(dst_yolo_path, 'w') as f:
                        f.write('\n'.join(yolo_lines))
                    annotated_file_path = dst_yolo_path
                
            except Exception as e:
                print(f"    ⚠ Error converting JSON for {img_filename}: {e}")
                skipped_count += 1
                continue
        
        # Priority 2: If JSON is missing, check for an existing YOLO TXT file and copy it
        elif os.path.exists(yolo_txt_source_path):
            try:
                # 🚨 COPY TXT DIRECTLY: Skip conversion if YOLO TXT is already present
                shutil.copyfile(yolo_txt_source_path, dst_yolo_path)
                annotated_file_path = dst_yolo_path
                
            except Exception as e:
                print(f"    ⚠ Error copying TXT for {img_filename}: {e}")
                skipped_count += 1
                continue
        
        else:
            # No annotation source found (neither JSON nor TXT)
            skipped_count += 1
            continue

        # --- FINAL STEP: COPY IMAGE AND UPDATE COUNT ---
        if annotated_file_path:
            shutil.copyfile(src_img, dst_img)
            
            if split == 'train':
                train_count += 1
            else:
                val_count += 1
        
print(f"\n✓ Dataset conversion complete!")
print(f"  - Train images: {train_count}")
print(f"  - Validation images: {val_count}")
print(f"  - Skipped (No valid annotation found): {skipped_count}")

# %%
```

```python
# %% [markdown]
# ## Step 18: Inference, Detailed Output, and Braille-to-English Conversion

# %%
import os
from IPython.display import display, Image as IPImage, Audio 
import pandas as pd
import numpy as np
PROJECT_ROOT = os.getcwd()
TEST_IMAGE_PATH = os.path.join(PROJECT_ROOT, "test_images/sample.jpg")
if trained_model is None:
    # If Step 17 failed, load the best model now for inference
    best_weights_path = os.path.join(results_dir, "weights", "best.pt")
    if os.path.exists(best_weights_path):
        trained_model = YOLO(best_weights_path)
    else:
        print("❌ Cannot proceed with inference: Best model weights not found.")
        # Raise an error or exit if the model is crucial
        # For notebook flow, we'll try to continue but flag the issue.
        # If this happens, ensure your training cell completed successfully first.
        trained_model = None
BRAILLE_TO_ENGLISH = {
    "1": "a", "12": "b", "14": "c", "145": "d", "15": "e", "124": "f",
    "1245": "g", "125": "h", "24": "i", "245": "j", "13": "k", "123": "l",
    "134": "m", "1345": "n", "135": "o", "1234": "p", "12345": "q", "1235": "r",
    "234": "s", "2345": "t", "136": "u", "1236": "v", "2456": "w", "1346": "x",
    "13456": "y", "1356": "z"
}
if trained_model is not None and os.path.exists(TEST_IMAGE_PATH):
    print("\n" + "=" * 60)
    print(f"🔍 RUNNING INFERENCE ON: {os.path.basename(TEST_IMAGE_PATH)}")
    print("=" * 60)

    inference_project_path = os.path.join(PROJECT_ROOT, "inference_output")
    
    # CRITICAL: Set device='cpu' for prediction (safe operation)
    predict_results = trained_model.predict(
        source=TEST_IMAGE_PATH, 
        conf=0.25, # Minimum confidence
        iou=0.3,   # Minimum IoU for NMS
        save=True, 
        project=inference_project_path, 
        name="final_prediction_output", # Changed name to avoid conflict with existing 'final_prediction'
        exist_ok=True,
        device='cpu' 
    )

    # --- 1. VISUALIZATION OF INFERENCE RESULTS ---
    # The saved image with detections
    inference_image_path = os.path.join(inference_project_path, "final_prediction_output", os.path.basename(TEST_IMAGE_PATH))
    
    if os.path.exists(inference_image_path):
        print("\n▼ Test Image with Detections:")
        display(IPImage(filename=inference_image_path, width=800))
    else:
        print("⚠️ Could not find visualized prediction image.")

    # --- 2. DETAILED DETECTION BREAKDOWN ---
    detection_data = []
    parsed_detections = []
    
    for r in predict_results:
        if r.boxes.cls.numel() > 0:
            for i in range(r.boxes.cls.numel()):
                cls_id = int(r.boxes.cls[i].item())
                conf = float(r.boxes.conf[i].item())
                xywh = r.boxes.xywh[i].cpu().numpy()
                
                braille_pattern = trained_model.names.get(cls_id, "?")
                english_char = BRAILLE_TO_ENGLISH.get(braille_pattern, "?")

                # Prepare data for translation logic (Step 18.2)
                parsed_detections.append([cls_id, float(xywh[0]), float(xywh[1])])
                
                # Prepare data for the detailed table
                detection_data.append({
                    "Confidence": f"{conf:.4f}",
                    "Class ID": cls_id,
                    "Braille Pattern": braille_pattern,
                    "English Character": english_char,
                    "X-Center": f"{xywh[0]:.2f}",
                    "Y-Center": f"{xywh[1]:.2f}"
                })

    print("\n✅ DETAILED DETECTION BREAKDOWN:")
    if detection_data:
        df_detections = pd.DataFrame(detection_data)
        # Sort by Y-Center (for line) then X-Center (for reading order)
        df_detections = df_detections.sort_values(by=["Y-Center", "X-Center"], ascending=[True, True])
        display(df_detections)
    else:
        print("⚠️ No detections found above the confidence threshold (0.25).")


    # --- 3. BRAILLE-TO-ENGLISH TRANSLATION ---
    # Logic is separated from display for cleanliness
    def detections_to_english_text(detections, class_map, y_threshold=30):
        """Sorts detections into lines (based on y-coordinate) and translates them."""
        # This function definition is critical for Step 18.3
        if not detections: return ""
        detections.sort(key=lambda d: (d[2], d[1]))
        lines, current_line = [], []
        if detections:
            current_line.append(detections[0])
            for det in detections[1:]:
                if abs(det[2] - current_line[-1][2]) < y_threshold:
                    current_line.append(det)
                else:
                    lines.append(sorted(current_line, key=lambda d: d[1]))
                    current_line = [det]
            lines.append(sorted(current_line, key=lambda d: d[1]))
        final_text_lines = []
        for line in lines:
            english_line = ""
            for det in line:
                class_id = det[0]
                braille_pattern = class_map.get(class_id, "?") 
                english_char = BRAILLE_TO_ENGLISH.get(braille_pattern, "?")
                english_line += english_char
            final_text_lines.append(english_line)
        return "\n".join(final_text_lines).strip()

    recognized_text = detections_to_english_text(parsed_detections, trained_model.names)
    
    print("\n✅ FINAL RECOGNIZED TEXT:")
    print("-" * 25)
    print(recognized_text)
    print("-" * 25)

else:
    if trained_model is None:
        print("❌ Inference skipped: Model was not loaded successfully.")
    else:
        print(f"❌ Inference skipped: Test image not found at '{TEST_IMAGE_PATH}'")

# %% [markdown]
# ## Step 19: Text-to-Speech Conversion and Audio Output

# %%
speech_file_path = os.path.join(PROJECT_ROOT, 'final_braille_speech.mp3')

if 'recognized_text' in locals() and recognized_text and recognized_text.strip():
    print("\n" + "=" * 60)
    print("🔊 GENERATING TEXT-TO-SPEECH")
    print("=" * 60)
    try:
        tts = gTTS(text=recognized_text, lang='en')
        tts.save(speech_file_path)
        print(f"✓ Speech saved to '{speech_file_path}'")
        
        # Display audio player in the notebook
        display(Audio(speech_file_path, autoplay=True))

    except Exception as e:
        print(f"❌ gTTS Failed: {e}")
        print("⚠️ Ensure you have an active internet connection to use gTTS.")
        
else:
    print("\n" + "=" * 60)
    print("⚠️ No valid text recognized. Skipping Text-to-Speech.")
    print("=" * 60)


print("\n🎉 ALL PIPELINE STEPS COMPLETE!")
# %%
```

```python
# %% [markdown]
# ## Step 19: Text-to-Speech Conversion, Tamil Translation, and Audio Output

# %%
from googletrans import Translator
from gtts import gTTS
from IPython.display import Audio

speech_file_path_en = os.path.join(PROJECT_ROOT, 'final_braille_speech_en.mp3')
speech_file_path_ta = os.path.join(PROJECT_ROOT, 'final_braille_speech_ta.mp3')

if 'recognized_text' in locals() and recognized_text and recognized_text.strip():
    print("\n" + "=" * 60)
    print("🔊 GENERATING TEXT-TO-SPEECH (English + Tamil)")
    print("=" * 60)
    try:
        # --- 1. English Speech ---
        tts_en = gTTS(text=recognized_text, lang='en')
        tts_en.save(speech_file_path_en)
        print(f"✓ English speech saved to '{speech_file_path_en}'")
        display(Audio(speech_file_path_en, autoplay=False))

        # --- 2. Tamil Translation ---
        translator = Translator()
        translated = translator.translate(recognized_text, src='en', dest='ta')
        tamil_text = translated.text
        print("\n🈶 Tamil Translation:")
        print("-" * 25)
        print(tamil_text)
        print("-" * 25)

        # --- 3. Tamil Speech ---
        tts_ta = gTTS(text=tamil_text, lang='ta')
        tts_ta.save(speech_file_path_ta)
        print(f"✓ Tamil speech saved to '{speech_file_path_ta}'")
        display(Audio(speech_file_path_ta, autoplay=False))

    except Exception as e:
        print(f"❌ Translation or gTTS Failed: {e}")
        print("⚠️ Ensure you have an active internet connection for Google Translate and gTTS.")
        
else:
    print("\n" + "=" * 60)
    print("⚠️ No valid text recognized. Skipping Text-to-Speech.")
    print("=" * 60)

print("\n🎉 ALL PIPELINE STEPS COMPLETE WITH TAMIL TRANSLATION!")
```
